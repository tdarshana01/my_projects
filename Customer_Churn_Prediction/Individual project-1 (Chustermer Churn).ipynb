# Model Building

```python
df2 = df1.drop('customerID',axis=1)
```

```python
df2['SeniorCitizen'] = df2['SeniorCitizen'].apply(lambda x:'Yes' if x == 1 else 'No')
```

```python
x = df2.drop('Churn',axis=1)
y = df2['Churn']
```

```python
y.value_counts()
```

```python
cat_features = [x for x in df2.columns if df2[x].dtype == 'object']
```

```python
cat_features = [x for x in cat_features if x != 'Churn']
```

```python
num_featuers = [x for x in df2.columns if df2[x].dtype != 'object']
```

```python
num_featuers
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from imblearn.pipeline import Pipeline
from sklearn.model_selection import StratifiedShuffleSplit, GridSearchCV, train_test_split
```

##### dividing original data into training and testing data

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

##### handle imbalance in the responce variable in the training set

```python
from imblearn.over_sampling import SMOTENC
smote = SMOTENC(categorical_features=cat_features,sampling_strategy='minority',random_state=42)
```

```python
X_train_sm, y_train_sm = smote.fit_resample(X_train,y_train)
```

##### encode categorical variables

```python
X_train_sm_encode = pd.get_dummies(X_train_sm)
```

```python
X_train_sm_encode = X_train_sm_encode.map(lambda x:1 if x == True else(0 if x == False else x))
```

```python
X_train_sm_encode.head()
```

```python
y_train_sm_encode = y_train_sm.apply(lambda x:1 if x == 'Yes' else 0)
```

```python
X_test_encode = pd.get_dummies(X_test)
```

```python
X_test_encode = X_test_encode.map(lambda x:1 if x == True else(0 if x == False else x))
```

```python
X_test_encode.head()
```

```python
y_test_encode = y_test.apply(lambda x:1 if x == 'Yes' else 0)
```

##### Scale numerical features

```python
preprocessor = ColumnTransformer([
    ('standardize',StandardScaler(),num_featuers)
],remainder='passthrough')
```

```python
X_train_sm_encode_scale = preprocessor.fit_transform(X_train_sm_encode)
```

```python
X_test_encode_scale = preprocessor.fit_transform(X_test_encode)
```

##### Train several machine learning models and evaluate them

```python
import numpy as np
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBRFClassifier
```

```python
models = [LogisticRegression(max_iter=1000),DecisionTreeClassifier(),RandomForestClassifier(),SVC(),KNeighborsClassifier(),XGBRFClassifier()]
```

```python
scores = []

for model in models:
    model.fit(X_train_sm_encode_scale,y_train_sm_encode)
    score = model.score(X_test_encode_scale,y_test_encode)
    score = np.round(score,2)

    scores.append({
        'model':model,
        'score':score
    })

pd.DataFrame(scores)
```

```python
def estimate_churn(gender,SeniorCitizen,Partner,Dependents,tenure,PhoneService,MultipleLines,InternetService,OnlineSecurity,OnlineBackup,DeviceProtection,TechSupport,StreamingTV,StreamingMovies,Contract,PaperlessBilling,PaymentMethod,MonthlyCharges,TotalCharges):

    X = np.zeros(len(x_train_1_prep.columns))
    fields = {
        'gender': gender,
        'Partner': Partner,
        'Dependents': Dependents,
        'SeniorCitizen': SeniorCitizen,
        'PhoneService': PhoneService,
        'MultipleLines': MultipleLines,
        'InternetService': InternetService,
        'OnlineSecurity': OnlineSecurity,
        'OnlineBackup': OnlineBackup,
        'DeviceProtection': DeviceProtection,
        'TechSupport': TechSupport,
        'StreamingTV': StreamingTV,
        'StreamingMovies': StreamingMovies,
        'Contract': Contract,
        'PaperlessBilling': PaperlessBilling,
        'PaymentMethod': PaymentMethod
    }

    for feature,value in fields.items():
        column_name = f"{feature}_{value}"
        if column_name in x_train_1_prep.columns:
            index = x_train_1_prep.columns.get_loc(column_name)
            X[index] = 1

    numerical_input = np.array([[tenure,MonthlyCharges,TotalCharges]])
    scaled_values = scaler.transform(numerical_input)[0]

    X[x_train_1_prep.columns.get_loc('tenure')] = scaled_values[0]
    X[x_train_1_prep.columns.get_loc('MonthlyCharges')] = scaled_values[1]
    X[x_train_1_prep.columns.get_loc('TotalCharges')] = scaled_values[2]

    prediction = model.predict([X])[0]
    result = ''
    if prediction == 1:
        result = 'Yes'
    else:
        result = 'No'

    probability = model.predict_proba([X])[0][1]
    return result,probability
```
